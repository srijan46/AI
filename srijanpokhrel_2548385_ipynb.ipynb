```python
# Task-1

south_asia_list = [
    "Afghanistan", "Bangladesh", "Bhutan", "India",
    "Maldives", "Nepal", "Pakistan", "Sri Lanka"
]

hdi_south_asia_df = df[df["country"].isin(south_asia_list)].copy()

hdi_south_asia_df.to_csv(
    "/content/drive/MyDrive/HDI_SouthAsia.csv",
    index=False
)
print("New dataset successfully created.")
```

```python
# Task 2

hdi_south_asia_df["Composite_Score"] = (
    0.30 * hdi_south_asia_df["life_expectancy"] +
    0.30 * hdi_south_asia_df["gross_inc_percap"]
)

country_scores = (
    hdi_south_asia_df.groupby("country", as_index=False)
    .agg({"Composite_Score": "mean", "hdi": "mean"})
)

sorted_composite = country_scores.sort_values(
    by="Composite_Score",
    ascending=False
)
top_5_countries = sorted_composite.head(5)

plt.figure(figsize=(8, 5))
plt.barh(top_5_countries["country"], top_5_countries["Composite_Score"])
plt.xlabel("Composite Development Score")
plt.ylabel("Country")
plt.title("Top 5 South Asian Countries by Composite Development Score")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
# Task 3

def identify_outliers(series):
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    return (series < lower_bound) | (series > upper_bound)

hdi_south_asia_df["HDI_outlier"] = identify_outliers(hdi_south_asia_df["hdi"])
hdi_south_asia_df["GNI_outlier"] = identify_outliers(hdi_south_asia_df["gross_inc_percap"])
hdi_south_asia_df["Outlier"] = hdi_south_asia_df["HDI_outlier"] | hdi_south_asia_df["GNI_outlier"]

plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=hdi_south_asia_df,
    x="gross_inc_percap",
    y="hdi",
    hue="Outlier",
    palette={False: "blue", True: "red"}
)
plt.title("GNI per Capita vs HDI (Outliers Highlighted)")
plt.xlabel("GNI per Capita")
plt.ylabel("HDI")
plt.ylim(0, 1)
plt.tight_layout()
plt.show()
```

```python
# Task 4

metrics_list = ["gender_development", "life_expectancy"]

for metric in metrics_list:
    correlation = df[metric].corr(df["hdi"])
    print(f"Pearson correlation between {metric} and HDI: {correlation:.3f}")

for metric in metrics_list:
    plt.figure(figsize=(6, 5))
    sns.regplot(
        data=df,
        x=metric,
        y="hdi",
        scatter_kws={"alpha": 0.6},
        line_kws={"color": "red"}
    )
    plt.title(f"HDI vs {metric.replace('_', ' ').title()}")
    plt.xlabel(metric.replace("_", " ").title())
    plt.ylabel("HDI")
    plt.ylim(0, 1)
    plt.tight_layout()
    plt.show()
```

```python
# 3
# Task-1

south_asia_list = [
    "Afghanistan", "Bangladesh", "Bhutan", "India",
    "Maldives", "Nepal", "Pakistan", "Sri Lanka"
]
middle_east_list = [
    "Bahrain", "Iran", "Iraq", "Israel", "Jordan",
    "Kuwait", "Lebanon", "Oman", "Palestine", "Qatar",
    "Saudi Arabia", "Syria", "United Arab Emirates", "Yemen"
]

hdi_south_asia_2020_22 = hdi_subset_years[hdi_subset_years["country"].isin(south_asia_list)].copy()
hdi_middle_east_2020_22 = hdi_subset_years[hdi_subset_years["country"].isin(middle_east_list)].copy()

hdi_south_asia_2020_22.to_csv(
    "/content/drive/MyDrive/HDI_SouthAsia_2020_2022.csv",
    index=False
)
hdi_middle_east_2020_22.to_csv(
    "/content/drive/MyDrive/HDI_MiddleEast_2020_2022.csv",
    index=False
)

print("South Asia and Middle East subsets saved successfully.")
```

```python
# Task-2

def calculate_region_stats(df, region_label):
    avg_hdi = df["hdi"].mean()
    hdi_std = df["hdi"].std()
    print(f"{region_label} - Mean HDI: {avg_hdi:.3f}, Std Dev: {hdi_std:.3f}")
    return avg_hdi

mean_south_asia = calculate_region_stats(hdi_south_asia_2020_22, "South Asia")
mean_middle_east = calculate_region_stats(hdi_middle_east_2020_22, "Middle East")

top_region = "South Asia" if mean_south_asia > mean_middle_east else "Middle East"
print(f"Region performing better on average: {top_region}")

```

```python
# Task-3

def get_top_bottom(df, region_label):
    country_avg = df.groupby("country", as_index=False)["hdi"].mean()
    top_3 = country_avg.sort_values(by="hdi", ascending=False).head(3)
    bottom_3 = country_avg.sort_values(by="hdi", ascending=True).head(3)
    display(f"{region_label} Top 3 :", top_3)
    display(f"{region_label} Bottom 3 :", bottom_3)
    return top_3, bottom_3

top_south_asia, bottom_south_asia = get_top_bottom(hdi_south_asia_2020_22, "South Asia")
top_middle_east, bottom_middle_east = get_top_bottom(hdi_middle_east_2020_22, "Middle East")

plt.figure(figsize=(10, 6))
combined_df = pd.concat([
    top_south_asia.assign(Region="South Asia"),
    top_middle_east.assign(Region="Middle East"),
    bottom_south_asia.assign(Region="South Asia"),
    bottom_middle_east.assign(Region="Middle East")
])

sns.barplot(
    data=combined_df,
    x="country",
    y="hdi",
    hue="Region"
)
plt.title("Top and Bottom Performers by HDI")
plt.ylabel("Average HDI")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# # Task 4

metrics = ["gender_development", "life_expectancy", "gross_inc_percap"]

sa_metrics = hdi_south_asia_2020_22[metrics].mean(numeric_only=True).to_frame().T.assign(Region="South Asia")
me_metrics = hdi_middle_east_2020_22[metrics].mean(numeric_only=True).to_frame().T.assign(Region="Middle East")

combined_melted = pd.concat([sa_metrics, me_metrics]).melt(id_vars="Region", var_name="Metric", value_name="Value")

plt.figure(figsize=(10,6))
sns.barplot(data=combined_melted, x="Metric", y="Value", hue="Region")
plt.title("Comparison of HDI Component Metrics Across Regions")
plt.ylabel("Average Value")
plt.tight_layout()
plt.show()
```

```python
# Task-5

def calculate_disparity(df, region_label):
    avg_hdi = df["hdi"].mean()
    std_hdi = df["hdi"].std()
    hdi_range = df["hdi"].max() - df["hdi"].min()
    coef_var = std_hdi / avg_hdi
    print(f"{region_label} - HDI Range: {hdi_range:.3f}, CV: {coef_var:.3f}")
    return coef_var

cv_south_asia = calculate_disparity(hdi_south_asia_2020_22, "South Asia")
cv_middle_east = calculate_disparity(hdi_middle_east_2020_22, "Middle East")

region_more_variation = "South Asia" if cv_south_asia > cv_middle_east else "Middle East"
print(f"Region with more variation in HDI: {region_more_variation}")
```

```python
# Task-6

metrics_to_check = ["gender_development", "life_expectancy"]

for region_label, df_region in [
    ("South Asia", hdi_south_asia_2020_22),
    ("Middle East", hdi_middle_east_2020_22)
]:
    print(f"\n{region_label} correlations:")
    for metric in metrics_to_check:
        correlation = df_region[metric].corr(df_region["hdi"])
        print(f"{metric} vs HDI: {correlation:.3f}")

        plt.figure(figsize=(6, 5))
        sns.regplot(
            data=df_region,
            x=metric,
            y="hdi",
            scatter_kws={"alpha": 0.6},
            line_kws={"color": "red"}
        )
        plt.title(f"{metric.replace('_', ' ').title()} vs HDI ({region_label})")
        plt.xlabel(metric.replace("_", " ").title())
        plt.ylabel("HDI")
        plt.ylim(0, 1)
        plt.tight_layout()
        plt.show()
```

```python
# Task-7

def identify_outliers(series):
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    return (series < lower_bound) | (series > upper_bound)

for region_label, df_region in [
    ("South Asia", hdi_south_asia_2020_22),
    ("Middle East", hdi_middle_east_2020_22)
]:
    df_copy = df_region.copy()
    df_copy["HDI_outlier"] = identify_outliers(df_copy["hdi"])
    df_copy["GNI_outlier"] = identify_outliers(df_copy["gross_inc_percap"])
    df_copy["Outlier"] = df_copy["HDI_outlier"] | df_copy["GNI_outlier"]

    print(f"\n{region_label} outliers:")
    display(df_copy[df_copy["Outlier"]][["country", "year", "hdi", "gross_inc_percap"]])

    plt.figure(figsize=(8, 6))
    sns.scatterplot(
        data=df_copy,
        x="gross_inc_percap",
        y="hdi",
        hue="Outlier",
        palette={False: "blue", True: "red"}
    )
    plt.title(f"GNI per Capita vs HDI Outliers ({region_label})")
    plt.xlabel("GNI per Capita")
    plt.ylabel("HDI")
    plt.ylim(0, 1)
    plt.tight_layout()
    plt.show()
```
