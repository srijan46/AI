```
import pandas as pd
import numpy as np

bank = pd.read_csv('bank.csv')
medical = pd.read_csv('medical_students_dataset.csv')
titanic = pd.read_csv('Titanic-Dataset.csv')
```

```
print("=== OBJECT COLUMNS — BANK ===")
display(bank.select_dtypes(include='object'))

print("=== OBJECT COLUMNS — MEDICAL ===")
display(medical.select_dtypes(include='object'))

print("=== OBJECT COLUMNS — TITANIC ===")
display(titanic.select_dtypes(include='object'))
```

```
print("=== MISSING VALUES — BANK ===")
display(bank.isnull().sum().to_frame("Missing Values"))

print("=== MISSING VALUES — MEDICAL ===")
display(medical.isnull().sum().to_frame("Missing Values"))

print("=== MISSING VALUES — TITANIC ===")
display(titanic.isnull().sum().to_frame("Missing Values"))
```

```
print("=== SUMMARY STATISTICS — BANK ===")
display(bank.describe().T)

print("=== SUMMARY STATISTICS — MEDICAL ===")
display(medical.describe().T)

print("=== SUMMARY STATISTICS — TITANIC ===")
display(titanic.describe().T)
```

```
bank_clean = bank.copy()
for col in bank_clean.columns:
    bank_clean[col].fillna(bank_clean[col].mode()[0], inplace=True)

print("BANK CLEANED ✓")
display(bank_clean.head())
```

```
medical_clean = medical.dropna()
print("MEDICAL CLEANED ")
display(medical_clean.head())
```

```
titanic_clean = titanic.copy()

titanic_clean['Age'].fillna(titanic_clean['Age'].mean(), inplace=True)
titanic_clean['Fare'].fillna(titanic_clean['Fare'].median(), inplace=True)

# "Embarked" missing → fill with mode
titanic_clean['Embarked'].fillna(titanic_clean['Embarked'].mode()[0], inplace=True)

print("TITANIC CLEANED ")
display(titanic_clean.head())
```

```
import matplotlib.pyplot as plt

plt.hist(titanic_clean['Age'])
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Titanic Age Distribution")
plt.show()
```

```
titanic_clean['Survived'].value_counts().plot(kind='bar')
plt.xlabel("Survived (0=No, 1=Yes)")
plt.ylabel("Count")
plt.title("Survival Count")
plt.show()
```

```
titanic_clean.boxplot(column='Fare')
plt.title("Fare Boxplot")
plt.show()
```

```
bank_clean.to_csv("bank_cleaned.csv", index=False)
medical_clean.to_csv("medical_cleaned.csv", index=False)
titanic_clean.to_csv("titanic_cleaned.csv", index=False)

print("All cleaned datasets saved successfully!")
```
