```python
import pandas as pd
import numpy as np
```

```python
titanic = pd.read_csv('Titanic-Dataset.csv')
```

```python
fare = titanic[['Fare']]
print("Fare DataFrame head:")
print(fare.head())
```

```python
class_age = titanic[['Pclass', 'Age']]
print("\nClass and Age DataFrame head:")
print(class_age.head())
```

```python
survived_gender = titanic[['Survived', 'Sex']]
print("\nSurvived and Gender DataFrame head:")
print(survived_gender.head())
```

```python
fare_gt_100 = titanic[titanic['Fare'] > 100]
print("Passengers with Fare > 100:")
print(fare_gt_100)
```

```python
first_class = titanic[titanic['Pclass'] == 1]
print("\nFirst class passengers:")
print(first_class)
```

```python
female_under_18 = titanic[(titanic['Age'] < 18) & (titanic['Sex'] == 'female')]
print("\nFemale passengers under 18:")
print(female_under_18)
```

```python
embarked_c_or_s = titanic[titanic['Embarked'].isin(['C', 'S'])]
print("Passengers embarked at Cherbourg (C) or Southampton (S):")
print(embarked_c_or_s)
```

```python
first_second_class = titanic[titanic['Pclass'].isin([1, 2])]
print("\nPassengers in first or second class:")
print(first_second_class)
```

```python
titanic['fare_per_year'] = titanic['Fare'] / titanic['Age']
```

```python
high_fare_age = titanic[titanic['fare_per_year'] > 5]
```

```python
high_fare_age_srt = high_fare_age.sort_values(by='fare_per_year', ascending=False)
```

```python
result = high_fare_age_srt[['Name', 'fare_per_year']]
```

```python
print(result)
```

```python
titanic['fare_per_class'] = titanic['Fare'] / titanic['Pclass']
adult_males = titanic[(titanic['Age'] >= 18) & (titanic['Sex'] == 'male')]

adult_males_srt = adult_males.sort_values(by='fare_per_class', ascending=False)

result = adult_males_srt[['Name', 'Age', 'fare_per_class']]
print(result)
```

```python
total_fare = titanic['Fare'].sum()
print(f"Total Fare (All Passengers): {total_fare:.2f}")

fare_p1 = titanic[titanic['Pclass'] == 1]['Fare'].sum()
print(f"Total Fare (First Class): {fare_p1:.2f}")

fare_p2 = titanic[titanic['Pclass'] == 2]['Fare'].sum()
print(f"Total Fare (Second Class): {fare_p2:.2f}")

fare_p3 = titanic[titanic['Pclass'] == 3]['Fare'].sum()
print(f"Total Fare (Third Class): {fare_p3:.2f}")

fare_totals = [fare_p1, fare_p2, fare_p3]

fare_percent = [(f / total_fare) * 100 for f in fare_totals]

for i, pct in enumerate(fare_percent, start=1):
    print(f"Passenger Class {i}: {pct:.2f}% of total fare revenue")
```

```python
def categorize_age(age):
    if age < 18:
        return 'child'
    elif 18 <= age < 65:
        return 'adult'
    else:
        return 'senior'

titanic['age_group'] = titanic['Age'].apply(categorize_age)

total_passengers = titanic.shape[0]
print(f"Total Passengers: {total_passengers}")

age_group_counts = titanic['age_group'].value_counts()
print("\nNumber of passengers by age group:")
print(age_group_counts)

age_group_percent = (age_group_counts / total_passengers) * 100

print("\nPercentage of passengers by age group:")
print(age_group_percent.round(2))
```
